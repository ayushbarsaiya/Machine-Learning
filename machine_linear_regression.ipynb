# data is not normally distributed, it is skewed than we apply iqr formula

```python
Max=df["MYCT"].max()
Min=df["MYCT"].min()
Mean=df["MYCT"].mean()
STD=df["MYCT"].std()
print("MAX: ",Max)
print("Min: ",Min)
print("Mean: ",Mean)
print("STD: ",STD)
```

# iqr formula,
# lower_limit=q1-1.5*iqr
# upper_limit=q3+1.5*iqr

```python
q1=df["MYCT"].quantile(0.25)
q3=df["MYCT"].quantile(0.75)
iqr=q3-q1
print("Q1: ",q1)
print("Q3: ",q3)
print("IQR: ",iqr)
```

```python
df["MYCT"].describe()
```

```python
lower_limit=q1-1.5*(iqr)
lower_limit
```

```python
upper_limit=q3+1.5*(iqr)
upper_limit
```

```python
plt.boxplot(df["MYCT"])
plt.show()
```

```python
MYCT_outlier=df[df["MYCT"]>upper_limit]
MYCT_outlier
```

```python
MYCT_outlier.shape[0]
```

# Linear Regression

```python
df.Model.nunique()
```

```python
df.drop(["Model"],axis=1,inplace=True)
```

```python
df
```

```python
df=pd.get_dummies(df)
```

```python
df
```

```python
X=df.drop(["ERP"],axis=1)
X
```

```python
y=df["ERP"]
```

# train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
X_train
```

```python
X_test.shape[0]
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
print("RMSE: ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
# model.score(y_test,y_pred)
model.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

# 92.86% accuracy of the model
